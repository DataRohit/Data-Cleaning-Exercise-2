# Importing Libraries

```python
# Data Manipulation
import pandas as pd

#ignore warning messages 
import warnings
warnings.filterwarnings('ignore')
```

# Loading Data

```python
dataset = pd.read_excel("Survey-Monkey-Data-Edited.xlsx", sheet_name="Edited_Data")
dataset
```

# Preprocessing

```python
dataset_modified = dataset.copy()
```

```python
dataset_modified.columns
```

```python
dataset_modified.drop(columns=['Start Date ', 'End Date ', 'Email Address ', 'First Name ', 'Last Name ', 'Custom Data 1 '], axis=1, inplace=True)
dataset_modified
```

## Applying Melt Method to Dataset

```python
id_vars = dataset_modified.columns[:8].to_list()
value_vars = dataset_modified.columns[8:].to_list()
```

```python
dataset_melted = dataset_modified.melt(id_vars=id_vars, value_vars=value_vars, var_name="Question_Subquestion", value_name="Answer")
dataset_melted
```

```python
dataset_melted["Question"] = dataset_melted["Question_Subquestion"].apply(lambda x: x.split("+")[0])

columns = dataset_melted.columns.to_list()
columns.remove("Question")
columns.insert(-2, "Question")

dataset_melted = dataset_melted[columns]
dataset_melted
```

```python
respondents = dataset_melted[dataset_melted["Answer"].notna()]
respondents = respondents.groupby(by="Question")["Respondent ID "].nunique().reset_index()
respondents.rename(columns={
    "Respondent ID ": "Respondents"
}, inplace=True)
respondents
```

```python
dataset_merged = pd.merge(left=dataset_melted, right=respondents, how="left", left_on="Question", right_on="Question")
dataset_merged
```

```python
same_answer = dataset_melted # [dataset_melted["Answer"].notna()]
same_answer = same_answer.groupby(by=["Question_Subquestion", "Answer"])["Respondent ID "].nunique().reset_index()
same_answer.rename(columns={
    "Respondent ID ": "Same answer"
}, inplace=True)
same_answer
```

```python
dataset_merged_two = pd.merge(left=dataset_merged, right=same_answer, how="left", left_on=["Question_Subquestion", "Answer"], right_on=["Question_Subquestion", "Answer"])
dataset_merged_two["Same answer"].fillna(0, inplace=True)
dataset_merged_two
```

```python
output = dataset_merged_two.copy()
output.rename(
    columns={
        'Identify which division you work in.+Response':"Division Primary",
        'Identify which division you work in.+Other (please specify)':"Division Secondary",
        'Which of the following best describes your position level?+Response': "Position",
        'Which generation are you apart of?+Response': "Generation",
        'Please select the gender in which you identify.+Response': "Gender",
        'Which duration range best aligns with your tenure at your company?+Response': "Tenure",
        'Which of the following best describes your employment type?+Response': "Employment Type"
    }, inplace=True
)
output
```

```python
output.to_excel("final_output.xlsx", index=False)
```
